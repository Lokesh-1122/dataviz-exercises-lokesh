**Scatter & Bubble Charts: Gapminder**

1.   Colour used sparingly — one categorical variable, no rainbow
2.   If showing all continents, either use accessible palette OR grey all + highlight one


1.   size_max set when using bubble size
2.   Log scale for GDP per capita


1.   Insight title












```python

import pandas as pd
import plotly.express as px


# Dataset: Gapminder — GDP, Life Expectancy, Population by Country
# Source: Gapminder Foundation (gapminder.org)

df = px.data.gapminder()
print(f"Loaded: {len(df)} rows")
print(df.head())
```

```python
# explore

print(df.info())
print("Years:", sorted(df['year'].unique()))
print("Continents:", df['continent'].unique())
print(df.describe().round(1))
```

**Task 1 — Scatter: life expectancy change over time**


**What to build:** A scatter showing GDP per capita vs life expectancy for two years (2000 and 2007) to show how both moved — use colour for year (just 2 colours), one continent only.



Choose any continent except Africa (that was the example). Highlight the change direction.



**💡 Filter: df.loc[df['continent'] == 'YOUR_CHOICE'] then filter years**

```python
# ---- FILTER ASIA  ----
continent = "Asia"
years = [2002, 2007]

df_sub = df[
    (df["continent"] == continent) &
    (df["year"].isin(years))
]

print(df_sub["year"].value_counts())  # MUST show both 2002 and 2007

# ---- SCATTER WITH STRONG DISTINCT COLOURS ----
fig = px.scatter(
    df_sub,
    x="gdpPercap",
    y="lifeExp",
    color="year",
    color_discrete_map={
        2002: "#ff7f0e",   # bright orange
        2007: "#1f77b4"    # deep blue
    },
    hover_name="country",
    size="pop",
    size_max=40,
    title="Asia: GDP per Capita vs Life Expectancy (2002 vs 2007)",
    labels={
        "gdpPercap": "GDP per Capita (USD)",
        "lifeExp": "Life Expectancy (Years)"
    }
)

fig.update_layout(template="plotly_white")
fig.show()
```

```python
continent = "Europe"
years = [1997, 2007]   # correct years

df_sub = df[
    (df["continent"] == continent) &
    (df["year"].isin(years))
]

fig = px.scatter(
    df_sub,
    x="gdpPercap",
    y="lifeExp",
    color="year",
    color_discrete_map={
        1997: "#1f77b4",   # blue
        2007: "#d62728"    # red
    },
    hover_name="country",
    size="pop",
    size_max=40,
    title=f"{continent}: GDP per Capita vs Life Expectancy (1997 vs 2007)",
    labels={
        "gdpPercap": "GDP per Capita (USD)",
        "lifeExp": "Life Expectancy (Years)"
    }
)

fig.update_layout(template="plotly_white")
fig.show()
```

**Task 2 — Bubble chart: tell a story**

**What to build:** A bubble chart (full 2007 dataset, all countries) where:

x = GDP per capita (log scale)

y = life expectancy

size = population

colour = ONE continent highlighted (your choice), all others grey

At least one annotation explaining the highlighted group's story

This is the grey-and-highlight technique applied to a bubble chart.

```python
import pandas as pd
import plotly.express as px

# Load dataset
df = px.data.gapminder()

# Filter for 2007 only
df_2007 = df[df["year"] == 2007]

# Choose continent to highlight
highlight = "Asia"

# Create a colour column: highlight vs grey
df_2007["color_group"] = df_2007["continent"].apply(
    lambda c: highlight if c == highlight else "Other"
)

# Custom colours
color_map = {
    highlight: "#e41a1c",   # strong red
    "Other": "#bdbdbd"      # grey
}

# Bubble chart
fig = px.scatter(
    df_2007,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="color_group",
    color_discrete_map=color_map,
    hover_name="country",
    size_max=60,
    log_x=True,
    title="Global Development in 2007: Highlighting Asia",
    labels={
        "gdpPercap": "GDP per Capita (log scale)",
        "lifeExp": "Life Expectancy (Years)",
        "color_group": "Continent"
    }
)

# Annotation explaining Asia's story
fig.add_annotation(
    x=12000,
    y=75,
    text="Asia shows wide diversity: from low-income countries<br>to rapidly growing economies like China & India.",
    showarrow=True,
    arrowhead=2,
    arrowsize=1,
    arrowcolor="#e41a1c",
    font=dict(size=12, color="#e41a1c")
)

fig.update_layout(template="plotly_white")
fig.show()
```
